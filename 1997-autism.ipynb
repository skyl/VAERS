```python
%matplotlib inline
import pandas as pd
from scipy import stats

data = pd.DataFrame.from_csv("./data/1997VAERSDATA.csv")
vax = pd.DataFrame.from_csv("./data/1997VAERSVAX.csv")
symptoms = pd.DataFrame.from_csv("./data/1997VAERSSYMPTOMS.csv")
```

```python
df = data.dropna(subset=["SYMPTOM_TEXT"])

vaxxers_idx = vax[vax.VAX_TYPE.str.contains("MMR")].index
vaxxers = df[df.index.isin(vaxxers_idx)]
non_vaxxers = df[~df.index.isin(vaxxers_idx)]

v = vaxxers[(vaxxers.AGE_YRS <= 1.6) & (vaxxers.AGE_YRS >= 0.6)]
nv = non_vaxxers[(non_vaxxers.AGE_YRS <= 1.6) & (non_vaxxers.AGE_YRS >= 0.6)]
```

```python
s = ( 
  df.SYMPTOM_TEXT.str.contains("autis", case=False) 
  | df.SYMPTOM_TEXT.str.contains("speech", case=False)
  | df.SYMPTOM_TEXT.str.contains("eye contact", case=False)
)
a = len(v.index.intersection(df[s].index))
b = len(v) - a
c = len(nv.index.intersection(df[s].index))
d = len(nv) - c

print(a, b, c, d)
oddsratio, pvalue = stats.fisher_exact([[a, b], [c, d]])
print(oddsratio, pvalue)
```
